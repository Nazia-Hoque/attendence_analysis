```python
import pandas as pd

# Load the CSV file
file_path = '2018-2019_Daily_Attendance_20240429.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Display basic information about the DataFram
df.info()

# Check for missing values
df.isnull().sum()

# Display basic statistics of the DataFrame
df.describe()

# Display the column names
df.columns
```

```python
# Calculate the mean, median, and standard deviation of relevant columns
mean_Enrolled = df['Enrolled'].mean()
median_Enrolled = df['Enrolled'].median()
std_Enrolled = df['Enrolled'].std()

print(f"Mean Enrolled: {mean_Enrolled}")
print(f"Median: Enrolled: {median_Enrolled}")
print(f"Standard Deviation of Enrolled : {std_Enrolled}")
```

```python
mean_Present = df['Present'].mean()
median_Present = df['Present'].median()
std_Present = df['Present'].std()
print(f"Mean Present: {mean_Present}")
print(f"Median: Present: {median_Present}")
print(f"Standard Deviation of Present : {std_Present}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot a histogram of the enrolled students
plt.figure(figsize=(10, 6))
sns.histplot(df['Enrolled'], bins=30, kde=True)
plt.title('Distribution of Enrolled Students')
plt.xlabel('Enrolled')
plt.ylabel('Frequency')
plt.show()

# Plot a histogram of the absent students
plt.figure(figsize=(10, 6))
sns.histplot(df['Absent'], bins=30, kde=True)
plt.title('Distribution of Absent Students')
plt.xlabel('Absent')
plt.ylabel('Frequency')
plt.show()

# Plot a histogram of the present students
plt.figure(figsize=(10, 6))
sns.histplot(df['Present'], bins=30, kde=True)
plt.title('Distribution of Present Students')
plt.xlabel('Present')
plt.ylabel('Frequency')
plt.show()

# Plot a time series of attendance (if the data has a date column)
plt.figure(figsize=(12, 6))
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')
sns.lineplot(x='Date', y='Present', data=df)
plt.title('Daily Attendance Over Time')
plt.xlabel('Date')
plt.ylabel('Present')
plt.show()

# Plot a boxplot to show the distribution of present students by month
df['Month'] = df['Date'].dt.month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Present', data=df)
plt.title('Attendance Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Present')
plt.show()
```
